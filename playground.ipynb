```python
from sentence_transformers import SentenceTransformer

# Load the MiniLM model (v6 or another variant)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example sentences
sentences = [
    "This is an example sentence.",
    "MiniLM is a smaller transformer model.",
    "Sentence embeddings are useful for many NLP tasks."
]

# Generate embeddings for the sentences
embeddings = model.encode(sentences)

# Display the embeddings for each sentence
for i, sentence in enumerate(sentences):
    print(f"Sentence: {sentence}")
    print(f"Embedding: {embeddings[i]}\n")
```
